```python
# O*NET-SOC Code:
# Official U.S. Standard Occupational Classification (SOC) code.
# Identifies the occupation uniquely.
# Example: 11-1011.00 = Chief Executives

# Title:
# Human-readable job title corresponding to the SOC code.

# Element ID:
# Hierarchical identifier for the worker characteristic.
# Indicates domain → category → specific ability.
# Mostly useful for taxonomy/hierarchy, not required for modeling.

# Element Name:
# Name of the ability being measured (e.g., Written Comprehension,
# Problem Sensitivity, Deductive Reasoning).
# This is the primary column used for ability filtering.

# Scale ID:
# Specifies the type of measurement.
# IM = Importance (how important the ability is for the job)
# LV = Level (how much of the ability is required)

# Scale Name:
# Text description of Scale ID.
# Redundant but useful for readability (Importance / Level).

# Data Value:
# The actual numeric score.
# Interpretation depends on Scale ID:
# - Importance (IM): 1–5 scale (Not important → Extremely important)
# - Level (LV): 1–7 scale (Low → Very high level required)

# N:
# Number of survey respondents who contributed to this rating.
# Smaller N implies higher uncertainty.

# Standard Error:
# Statistical uncertainty of the Data Value.
# Lower values indicate more reliable estimates.
# Can be used for confidence weighting.

# Lower CI Bound:
# Lower bound of the 95% confidence interval for the Data Value.

# Upper CI Bound:
# Upper bound of the 95% confidence interval for the Data Value.

# Recommend Suppress:
# Indicates whether O*NET recommends suppressing this value.
# 'Y' = do not use/display
# 'N' = safe to use
# Recommended: filter out rows with 'Y'.

# Not Relevant:
# Indicates whether the ability is considered irrelevant for the occupation.
# 'Y' = not relevant (should be excluded)
# 'N' or NaN = relevant

# Date:
# Date when the data point was last updated (MM/YYYY).
# Useful for recency checks or longitudinal analysis.

# Domain Source:
# Source of the rating (e.g., Analyst, Incumbent, Occupational Expert).
# Analysts = expert judgment, not self-report.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def rank_all_users_for_job(job_name, pa_new, ja_pivot, atwa_matrix, wa_final,
                           top_n=10, top_k_activities=3):

    # resolve job
    resolved_job = find_best_job(job_name, ja_pivot.index)

    abilities = ja_pivot.columns

    users = pa_new[abilities]

    # normalize user abilities
    means = pa_new[abilities].mean(skipna=True)
    stds = pa_new[abilities].std(skipna=True)

    users_norm = ((users - means) / stds).fillna(0)

    # job ability vector
    job_vector = ja_pivot.loc[resolved_job].values.reshape(1, -1)

    # similarity scores
    scores = cosine_similarity(users_norm.values, job_vector).flatten()

    results = pd.DataFrame({
        "user_id": pa_new.index,
        "match_score": scores
    })

    results["match_percent"] = (results["match_score"] + 1) / 2 * 100

    results = results.sort_values("match_score", ascending=False)

    if top_n:
        results = results.head(top_n)

    # -------- Activities explanation --------

    atwa_filtered = atwa_matrix.loc[abilities]

    job_activity = wa_final.loc[resolved_job]

    activity_explanations = []

    for uid in results["user_id"]:

        user_vector = users_norm.loc[uid]

        # abilities → activities
        user_activity = atwa_filtered.T.dot(user_vector)

        # overlap with job activities
        overlap = user_activity * job_activity

        top_activities = (
            overlap.sort_values(ascending=False)
            .head(top_k_activities)
            .index
            .tolist()
        )

        activity_explanations.append(top_activities)

    results["SupportingActivities"] = activity_explanations

    return results
```

```python

from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def predict_jobs_ability_only(raw_scores, ja_pivot, top_n=10, top_k_abilities=3):

    abilities = ja_pivot.columns

    user_vector = pd.Series(raw_scores).reindex(abilities).fillna(0)

    user_array = user_vector.values.reshape(1, -1)

    ability_sim = cosine_similarity(user_array, ja_pivot.values)[0]

    results = pd.DataFrame({
        "Job": ja_pivot.index,
        "AbilitySimilarity": ability_sim
    })

    results = results.sort_values("AbilitySimilarity", ascending=False).head(top_n)

    explanations = []

    for job in results["Job"]:

        job_abilities = ja_pivot.loc[job]

        overlap = abs(user_vector) * abs(job_abilities)

        top_abilities = overlap.sort_values(ascending=False).head(top_k_abilities).index.tolist()

        explanations.append(top_abilities)

    results["MatchedAbilities"] = explanations

    return results
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from rapidfuzz import process, fuzz
import pandas as pd

def find_best_job(job_name, job_index):

    job_name = job_name.lower()

    # 1️⃣ exact match
    for job in job_index:
        if job.lower() == job_name:
            return job

    # 2️⃣ partial word match
    for job in job_index:
        if job_name in job.lower():
            return job

    # 3️⃣ fuzzy fallback
    match = process.extractOne(job_name, job_index, scorer=fuzz.token_sort_ratio)
    return match[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

def job_match(job_name, raw_scores, pa_new, ja_pivot, atwa_matrix, wa_final,
              top_k_strengths=5, top_k_weaknesses=5):

    # resolve job
    closest_job = find_best_job(job_name, wa_final.index)

    abilities = ja_pivot.columns

    # align user scores
    user_raw = pd.Series(raw_scores).reindex(abilities)

    means = pa_new[abilities].mean()
    stds = pa_new[abilities].std()

    user_vector = ((user_raw - means) / stds).replace([np.inf, -np.inf], 0).fillna(0)

    # ---- align ability → activity matrix ----
    atwa_filtered = atwa_matrix.reindex(abilities).fillna(0)

    # compute user activity scores
    user_activity = atwa_filtered.T.dot(user_vector)

    # ---- align job activity profile ----
    job_activity = wa_final.loc[closest_job].fillna(0)

    common_activities = user_activity.index.intersection(job_activity.index)

    user_activity = user_activity.loc[common_activities]
    job_activity = job_activity.loc[common_activities]

    # similarity score
    score = cosine_similarity(
        user_activity.values.reshape(1, -1),
        job_activity.values.reshape(1, -1)
    )[0][0]

    match_percent = (score + 1) / 2 * 100

    # strengths
    activity_strength = user_activity * job_activity

    strengths = (
        activity_strength
        .sort_values(ascending=False)
        .head(top_k_strengths)
        .index
        .tolist()
    )

    # weaknesses
    activity_gap = job_activity - user_activity

    weaknesses = (
        activity_gap
        .sort_values(ascending=False)
        .head(top_k_weaknesses)
        .index
        .tolist()
    )

    return {
        "job": closest_job,
        "match_percent": round(match_percent, 2),
        "strength_activities": strengths,
        "weakness_activities": weaknesses
    }
```

## func 4 : Estimate which work activities the user would perform well in based on their abilities.

```python
def infer_work_context_strengths(user_abilities, atwc):
    """
    Predict work contexts a user would perform well in based on tested abilities.

    Parameters
    ----------
    user_abilities : dict or pandas Series
        ability_name -> score
    atwc : DataFrame
        mapping table with:
        'Abilities Element Name'
        'Work Context Element Name'

    Returns
    -------
    pandas Series
        ranked work context strengths
    """

    import pandas as pd

    # convert to Series
    user_abilities = pd.Series(user_abilities)

    # abilities actually measured by the test
    tested_abilities = [
        "Deductive Reasoning",
        "Mathematical Reasoning",
        "Memorization",
        "Perceptual Speed",
        "Problem Sensitivity",
        "Selective Attention",
        "Speed of Closure",
        "Time Sharing",
        "Written Comprehension"
    ]

    # keep only tested abilities
    user_abilities = user_abilities[user_abilities.index.isin(tested_abilities)]

    # optionally remove abilities with score 0
    user_abilities = user_abilities[user_abilities != 0]

    # filter mapping table
    relevant = atwc[
        atwc["Abilities Element Name"].isin(user_abilities.index)
    ]

    # attach ability scores
    relevant = relevant.assign(
        ability_score=relevant["Abilities Element Name"].map(user_abilities)
    )

    # aggregate context strengths
    context_scores = (
        relevant
        .groupby("Work Activities Element Name")["ability_score"]
        .sum()
        .sort_values(ascending=False)
    )

    return context_scores
```

## func 5: Convert a job into its activity profile using O*NET data.

```python
def get_job_activity_profile(job_name, wa_final, find_best_job, top_n=10):
    """
    Return the top ranked work activities for a job.

    Parameters
    ----------
    job_name : str
        user input job name
    wa_final : DataFrame
        job × activity matrix
    find_best_job : function
        fuzzy job matching function
    top_n : int
        number of top activities to return

    Returns
    -------
    pandas Series
        ranked work activities
    """

    # resolve job name
    matched_job = find_best_job(job_name, wa_final.index)

    # activity vector
    activity_profile = wa_final.loc[matched_job]

    # rank activities
    ranked_activities = (
        activity_profile
        .sort_values(ascending=False)
        .head(top_n)
    )

    return ranked_activities
```

## func 8: Skill Gap Analysis (Very Strong Feature
skill gap
+
activity importance

```python
import pandas as pd
import numpy as np

def skill_gap_analysis(job_name, raw_scores, pa_new, ja_pivot, atwa_matrix, wa_final, top_k=5):

    job = find_best_job(job_name, wa_final.index)

    abilities = ja_pivot.columns

    # ----- normalize user abilities -----
    user_raw = pd.Series(raw_scores).reindex(abilities)

    means = pa_new[abilities].mean()
    stds = pa_new[abilities].std()

    user_vector = ((user_raw - means) / stds).replace([np.inf, -np.inf], 0).fillna(0)

    # ----- abilities → activities -----
    atwa_filtered = atwa_matrix.loc[abilities]

    user_activity = atwa_filtered.T.dot(user_vector)

    # normalize user activity profile
    user_activity = user_activity / (user_activity.abs().max() + 1e-6)

    # ----- job activity profile -----
    job_activity = wa_final.loc[job].fillna(0)

    # normalize job activity importance
    job_activity = job_activity / (job_activity.abs().max() + 1e-6)

    # align activities
    common = user_activity.index.intersection(job_activity.index)

    user_activity = user_activity.loc[common]
    job_activity = job_activity.loc[common]

    # ----- hybrid gap score -----
    gap = job_activity * (job_activity - user_activity)

    gap = gap.sort_values(ascending=False)

    top_gaps = gap.head(top_k)

    return {
        "job": job,
        "skill_gaps": pd.DataFrame({
            "Activity": top_gaps.index,
            "GapScore": top_gaps.values,
            "JobImportance": job_activity.loc[top_gaps.index].values,
            "UserCapability": user_activity.loc[top_gaps.index].values
        })
    }
```
